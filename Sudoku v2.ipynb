```python
class Cell:
    def __init__(self,x,y,b):
        self.x = x # x co-ordinate within a matrix
        self.y = y # y co-ordinate within a matrix
        self.box = b # in a Sudoko, there are 9 boxes of 9 cells. box represents these boxes with a val from 1-9
        self.given = 0 # given numbers. Will default to 0 if no given number for this cell.
        self.guess = 0 # track numbers to be guessed, defaulted to 0.
    
    def print_XY(self):
        print(self.x,":",self.y,":box ",self.box,", given:",self.given)
        
    def get_box(self):
        return self.box
    
    def get_x(self):
        return self.x
    
    def get_y(self):
        return self.y
    
    def get_given(self):
        return self.given
    

# Dictionary for matrix of Cells
sudoku = {}    

# Assign a box, 0-8, based on the x, y co-ordinates of the Cell
def assign_box(x,y):
    # top row of boxes:
    if y < 3 and x <3:
        return 0
    elif y < 3 and x < 6:
        return 1
    elif y < 3 and x < 9:
        return 2
    # second row of boxes:
    elif y < 6 and x <3:
        return 3
    elif y < 6 and x < 6:
        return 4
    elif y < 6 and x < 9:
        return 5
    # 3rd row of boxes
    elif y < 9 and x <3:
        return 6
    elif y < 9 and x < 6:
        return 7
    elif y< 9 and x < 9:
        return 8
    else:
        print("ERROR: check ",x,y)
    
# Generate sudoku cells
pos_counter = 0
for y in range(9):
    for x in range(9):
        sudoku[pos_counter] = Cell(x,y,assign_box(x,y))
        pos_counter += 1
```

```python
given_nums = [0,0,0,0,7,0,0,3,4
            ,2,7,0,6,0,0,5,0,0
            ,4,0,0,2,5,3,0,7,1
            ,0,6,0,0,9,0,0,0,0
            ,9,0,0,8,0,4,0,0,2
            ,0,0,0,0,1,0,0,6,0
            ,6,1,0,5,2,9,0,0,7
            ,0,0,8,0,0,6,0,4,5
            ,3,9,0,0,8,0,0,0,0]

#set the given numbers for our Sudoku cell objects
for n in sudoku:
    setattr(sudoku[n],'given',given_nums[n])

```

```python
# For a Cell attribute 'n', create a list to test. Available attributes to test are "y", "x", "box"
def test_attr(attr,n):    
    dup_counter = 0
    row = []
    for key in sudoku:
        if getattr(sudoku[key],str(attr)) == n and (getattr(sudoku[key],"given")) != 0: # append given values which are not 0
            row.append(getattr(sudoku[key],"given"))
        elif getattr(sudoku[key],str(attr)) == n and (getattr(sudoku[key],"guess")) != 0: # append guessed values which are not 0 
            row.append(getattr(sudoku[key],"guess"))                
    print(row)
    if len(set(row)) != len(row): # duplicate values are removed in sets
        dup_counter += 1
    return dup_counter #counter with 0 means that there are duplicate values in row, so need to run backtrack function.

    
```
